```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
tot = purchase_data["Gender"].value_counts()
conv = pd.DataFrame(tot)
col_rename_dict = {
    "Gender": "Total Count"
}
renamed_df = conv.rename(columns=col_rename_dict)
def compute_percentage(x):
    pct = (float(x)/renamed_df["Total Count"].sum())*100
    return round(pct,2)
renamed_df["Percentage"] = renamed_df.apply(compute_percentage, axis = 1)
renamed_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
count = purchase_data.groupby("Gender").count()
count = count["Purchase ID"]
average = purchase_data.groupby("Gender").mean()
average = round(average["Price"],2)
total = purchase_data.groupby("Gender").sum()
total = total["Price"]
res = {
    "Purchase Count" : count,
    "Avg Purchase Price" : average,
    "Avg Purchase Total" : total
}
sum_df = pd.DataFrame(res)
sum_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python

# Create bins in which to place values based upon age based bins
bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]
# Create labels for these bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29",
                "30-34", "35-39", "40+"]
s = pd.DataFrame(purchase_data.groupby(pd.cut(purchase_data["Age"], bins, labels=group_labels)).size())
rename_column = {
    0 : "Total Count"
}
s = s.rename(columns = rename_column)
def calc_percentage (x):
    pn = (float(x)/purchase_data["SN"].count()*100)
    return round(pn,2)
s["Percentage of Players"] = s.apply(calc_percentage, axis =1)
s
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create bins in which to place values based upon age based bins
bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]
# Create labels for these bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29",
                "30-34", "35-39", "40+"]
n = purchase_data.groupby(pd.cut(purchase_data["Age"], bins = bins, labels=group_labels)).count()
n1 = n["Purchase ID"]
t = purchase_data.groupby(pd.cut(purchase_data["Age"], bins = bins, labels=group_labels)).sum()
t1 = t["Price"]
a = purchase_data.groupby(pd.cut(purchase_data["Age"], bins = bins, labels=group_labels)).mean()
a1 = a["Price"]
purchasing_analysis = {
    "Purchase Count" : n1,
    "Avg Purchase Price" : round(a1,2),
    "Total Purchase Value" : t1
}
pa = pd.DataFrame(purchasing_analysis)
pa
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
columns = ["Item ID", "Item Name", "Price"]
pop_items = purchase_data[columns]
purchase_count = pop_items["Price"]
pop_items["purchase_count"] = purchase_count
li =["Item ID", "Item Name", "Price"]
pop_items = pop_items.groupby(li, as_index = False)["purchase_count"].count()
pop_items = pd.DataFrame(pop_items)
pop_items = pop_items.sort_values(by = "purchase_count", ascending = False)
totale = (pop_items["Price"] * pop_items["purchase_count"])
pop_items["Total"] = totale
pi =["Item ID", "Item Name"]
pop_items = pop_items.set_index(keys = pi)
pop_items
```
